```python
#Find Elements by ID
import requests
from bs4 import BeautifulSoup

URL = "https://www.christuniversity.in"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find(id="mega-menu-25293-0")
print(results.prettify())  #Prettify - creates a parse tree for parsed pages that can be used to extract data from HTML
```

```python
#Find Elements by HTML Class Name
import requests
from bs4 import BeautifulSoup

URL = "https://www.christuniversity.in"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
a=soup.find( class_ = "mega-menu-row mega-singlecolumnmobile" )
print(a)
```

```python
import requests
from bs4 import BeautifulSoup


# Making a GET request
r = requests.get('https://www.christuniversity.in')

# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')

# finding element by id
s = soup.find('div', class_= 'to-p-hde-r')

content = s.find_all('li')
print(content)
```

```python
#Extracting Text from the tags
import requests
from bs4 import BeautifulSoup


# Making a GET request
r = requests.get('https://www.christuniversity.in')

# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')

# finding element by id
s = soup.find('div', class_= 'to-p-hde-r')

lines = s.find_all('a')

for line in lines:
  print(line.text)
```
